```python
a=np.array([[1, -6, -10],
       [2, 9, 5],
       [0, -4, 12]])
eigenvalues, eigenvectors= np.linalg.eig(a)
```

```python
print("Власні числа:")
print(eigenvalues)
```

```python
print("\nВласні вектори:")
print(eigenvectors)
```
