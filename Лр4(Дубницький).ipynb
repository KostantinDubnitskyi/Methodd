```python
df=df.fillna(df.mean(numeric_only=True), axis=0)
```

```python
df.columns= df.columns.str.lower().str.replace(' ','_').str.replace('/','_').str.replace('__','_').str.strip()
df.columns
```

```python
df.tail()
```

```python
df['month'] = pd.to_datetime(df['month'], format = '%m/%d/%Y %I:%M:%S %p')

df_2018=df[df['month'].dt.year ==2018]

min_visitors = df_2018.iloc[:, 1:].min()
max_visitors = df_2018.iloc[:, 1:].max()
result=pd.DataFrame({'min': min_visitors, 'max' : max_visitors})

result
```

```python
average_visitors = df.iloc[:, 1:].mean()
average_visitors
```

```python
df_2015=df.loc[df['month'].dt.year ==2015].copy()

df_2015['total_visitors']=df_2015.iloc[:, 1:].sum(axis=1)

min_row = df_2015.loc[df_2015['total_visitors'].idxmin()]
max_row = df_2015.loc[df_2015['total_visitors'].idxmax()]

min_month=min_row['month'].strftime('%B')
max_month=max_row['month'].strftime('%B')

print(f'Найнижча відвідуваність у {min_month}')
print(f'Найвища відвідуваність у {max_month}')
```

```python
df_2015.tail()
```

```python
df_2017 = df[df['month'].dt.year == 2017].copy()

df_2017['month_num']=df_2017['month'].dt.month

autumn=df_2017[df_2017['month_num'].isin([9, 10, 11])]['avila_adobe'].sum()

spring=df_2017[df_2017['month_num'].isin([3, 4, 5])]['avila_adobe'].sum()

print(f'Людей відвідали {autumn} осінью')
print(f'Людей відвідали {spring} весною ')
```

```python
df.corr()
```

```python
mtx =df.corr(numeric_only=True).abs()

fig, ax = plt.subplots(figsize=(10,10))

sns.heatmap(
    mtx,
    cmap='crest',
    annot=True,
    fmt='.2f',
    linewidths=1,
    mask=np.triu(np.ones_like(mtx, dtype=bool)),
    square=True,
    cbar=False,
    ax=ax
)

plt.title("Матриця попарної кореляції ознак", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.lineplot(data=df , x='month' , y='avila_adobe')
plt.title ('Avila Adobe: Відвідуваність з часом')
plt.show()
```

```python
df['year'] = df['month'].dt.year
sns.boxplot(data=df, x='year' , y='chinese_american_museum')
plt.title('Chinese American Museum: Розподіл по роках')
plt.show()
```

```python
sns.histplot(df['gateway_to_nature_center'], bins=30, kde=True)

plt.title('Розподіл відвідуванності Gateway to Nature Center')
plt.show()
```

```python
avg_visitor=df.drop(columns=['month', 'year']).mean().sort_values(ascending=False)
sns.barplot(x=avg_visitor.values, y=avg_visitor.index)
plt.title('Середня кількість відвідувачів по музеях')
plt.xlabel('Середне значення')
plt.ylabel('Музей')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('penguins')
df.head()
```

```python
sns.scatterplot(data=df, x='bill_length_mm', y='bill_depth_mm', hue='species')

plt.title('Залежність довжини і глибини дзьоба')
plt.show()
```

```python
sns.countplot(data=df, x='species')

plt.title('Кількість пінгвінів по видам')
plt.show()
```

```python
sns.kdeplot(data=df, x='body_mass_g', hue='species', fill=True)

plt.title('Розподіл маси тіла')
plt.show()
```

```python
sns.stripplot(data=df, x='species', y='bill_length_mm', hue='sex', dodge=True)

plt.title('Довжина дзьоба по видам і статі')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='species', y='flipper_length_mm', ci='sd', palette='pastel')
plt.title('Середня довжина плавника по видах')
plt.ylabel('Довжина плавника (мм)')
plt.show()
```
