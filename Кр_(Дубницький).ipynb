1. Створіть горизонтальну діаграму з мітками на основі даних про врожайність пшениці (data.wheat()) за допомогою бібліотеки Altair. Задайте вісь х для врожайності пшениці, вісь у для років, а текстові мітки відображайте значення врожайності пшениці на відповідних роках.

```python
import altair as alt
from vega_datasets import data

source = data.wheat()

chart1 = alt.Chart(source).mark_bar().encode(
    x='wheat:Q',
    y='year:O',
    text='wheat:Q'
).properties(
    title="Врожайність пшениці"
)

chart1
```

2. Створіть багатошарову діаграму на основі даних про виробництво електроенергії в штаті Айова (data.iowa_electricity()) за допомогою бібліотеки Altair. Вісь х представляє роки, вісь у відображає чисте виробництво електроенергії (net generation), а кожен шар діаграми позначає різні джерела енергії (source). Встановіть прозорість (opacity) для стовпців на рівні 0.7.

```python
source = data.iowa_electricity()

chart2 = alt.Chart(source).mark_bar(opacity=0.7).encode(
    x='year:O',
    y='net_generation:Q',
    color='source:N'
).properties(
    title="Виробництво електроенергії в Айові"
)

chart2
```

3. Створіть графік за допомогою бібліотеки Plotly (scatter plot) для даних ірисів (iris dataset), де:
Вісь Х (х-ахія) відповідає ширині чашелистика (sepal width).
Вісь у (у-ахія) відповідає довжині чашелистика (sepal length). Кожен пункт на графіку має розмір, який залежить від довжини пелюстки (petal length).
Кожен пункт на графіку має різний колір в залежності від виду (species).
При наведенні на пункт повинні відображатися додаткові дані, такі як ширина пелюстки

```python
import plotly.express as px

df = px.data.iris()

fig = px.scatter(
    df,
    x='sepal_width',
    y='sepal_length',
    size='petal_length',
    color='species',
    hover_data=['petal_width'],
    title="Iris scatter plot"
)

fig.show()
```
