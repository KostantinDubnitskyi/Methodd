```python
import matplotlib.pyplot as plt
import numpy as np

x, y = 0, 0
angle = 0

xs = [x]
ys = [y]

length = 200

while length > 5:
    # рух вперед
    x += length * np.cos(np.radians(angle))
    y += length * np.sin(np.radians(angle))

    xs.append(x)
    ys.append(y)

    # поворот
    angle += 90

    # зменшення довжини
    length -= 5

plt.plot(xs, ys)
plt.gca().set_aspect('equal')
plt.title("Спіраль")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def generate(axiom, rules, iterations):
    result = axiom
    for _ in range(iterations):
        new_result = ""
        for char in result:
            new_result += rules.get(char, char)
        result = new_result
    return result

def draw(path, angle):
    x, y = 0, 0
    direction = 0

    stack = []

    xs = [x]
    ys = [y]

    step = 1

    for command in path:
        if command == "F":
            x += step * np.cos(np.radians(direction))
            y += step * np.sin(np.radians(direction))
            xs.append(x)
            ys.append(y)

        elif command == "+":
            direction += angle

        elif command == "-":
            direction -= angle

        elif command == "[":
            stack.append((x, y, direction))

        elif command == "]":
            x, y, direction = stack.pop()
            xs.append(x)
            ys.append(y)

    plt.plot(xs, ys)
    plt.gca().set_aspect('equal')
    plt.title("L-система фрактал")
    plt.show()


axiom = "F"
rules = {
    "F": "+F--F+"
}
angle = 45

path = generate(axiom, rules, 10)
draw(path, angle)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def generate(axiom, rules, iterations):
    result = axiom
    for _ in range(iterations):
        new_result = ""
        for char in result:
            new_result += rules.get(char, char)
        result = new_result
    return result

def draw_dragon(instructions, angle, step=1):
    x, y = 0, 0
    direction = 0

    xs = [x]
    ys = [y]

    for cmd in instructions:
        if cmd == "F":
            x += step * np.cos(np.radians(direction))
            y += step * np.sin(np.radians(direction))
            xs.append(x)
            ys.append(y)

        elif cmd == "+":
            direction -= angle

        elif cmd == "-":
            direction += angle

    plt.plot(xs, ys)
    plt.gca().set_aspect('equal')
    plt.title("Крива Дракона (L-система)")
    plt.show()


axiom = "FX"
rules = {
    "X": "X+YF+",
    "Y": "-FX-Y"
}
angle = 90

path = generate(axiom, rules, 10)
draw_dragon(path, angle)
```
